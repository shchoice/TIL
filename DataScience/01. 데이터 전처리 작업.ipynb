# 데이터 전처리 작업

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df
```

## 1. 특정 열 추출

### 특정 열 추출

```python
## (sepal length (cm))열 추출
print(df['sepal length (cm)'])
print('-' * 50)

## (모든 행, 3번째 열)
print(df.iloc[:, 3]) 
```

### 특정 행 추출

```python
# target이 0인 것을 만족하는 모든 행 추출
sample1 = df[df['target'] == 0]
display(sample1)
print()

# 3번째 행의 모든 열 출력
display(df.iloc[3,])
```

## 2. 열 또는 행에 함수를 적용
* apply 함수는 데이터프레임의 열 또는 행에 함수를 적용하는 데 사용됨
* 함수는 주어진 함수를 데이터프레임의 각 열 또는 각 행에 적용하고 결과를 반환

```python
def getLengthSize(x):
    if x > 6:
        return 'L'
    else:
        return 'S'
    
df['LenSize'] = df['sepal length (cm)'].apply(getLengthSize)
display(df)
```

## 3. pandas를 활용한 One-hot Encoding

```python
df2 = pd.DataFrame({'Species': ['Versicolor', 'Setosa', 'Virginica', 'Setosa', 'Setosa']})
encoded_df = pd.get_dummies(df2['Species'])
display(encoded_df)

encoded_df = pd.get_dummies(df2['Species'], drop_first=True)
display(encoded_df)
```
